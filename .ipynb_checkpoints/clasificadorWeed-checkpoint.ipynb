```python
train=[]
dev=[]
test=[]
tamano_img=100
for i, (img, label) in enumerate(datos['train']):
    if i<=12256:
        img=cv2.resize(img.numpy(),(tamano_img,tamano_img))
        img=img.reshape(tamano_img,tamano_img,3)
        train.append([img,label])
    elif 12256<i and i<=15758:
        img=cv2.resize(img.numpy(),(tamano_img,tamano_img))
        img=img.reshape(tamano_img,tamano_img,3)
        dev.append([img,label])
    else:
        img=cv2.resize(img.numpy(),(tamano_img,tamano_img))
        img=img.reshape(tamano_img,tamano_img,3)
        test.append([img,label])    
```

```python
x=[] #imagenes de entrada
y=[] #labels

x_dev=[] #imagenes de entrada
y_dev=[] #labels

x_test=[] #imagenes de entrada
y_test=[] #labels
for img, label in train:
    x.append(img)
    y.append(label)
for img, label in dev:
    x_dev.append(img)
    y_dev.append(label)
for img, label in test:
    x_test.append(img)
    y_test.append(label)
```
